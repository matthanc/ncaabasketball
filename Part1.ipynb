Create a neural network that can convert Celsius to Fahrenheit and examine the weights.
1. Using TensorFlow and Keras build a single layered network with a single input and a single output and a single neuron.
2. Using NumPy generate ~20 input/output values to test
3. Train the neural network on your data.
4. Test the neural network to see if it is working.
5. Examine the weights to see if they match F = C * 1.8 + 32F=Câˆ—1.8+32.

```
tempdf = pd.DataFrame()
tempdf['Celsius'] = np.random.randint(0,60,20)
tempdf['Farenheit'] = tempdf['Celsius'] * 1.8 + 32
xtrain = tempdf['Celsius']
ytrain = tempdf['Farenheit']
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1, input_shape = [1]))
model.compile(optimizer = tf.keras.optimizers.Adam(1), loss = 'mean_squared_error')
epochs_hist = model.fit(xtrain, ytrain, epochs = 500)
```

```
temp_C = 48
temp_F = model.predict([temp_C])
print(temp_C, 'celsius converted to fahrenheit is', temp_F)
```
